```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('device_average_inference_time_list_dict.json', 'r') as f:
    device_average_inference_time_list_dict = json.load(f)
with open('constellation_running_time_list_dict.json', 'r') as f:
    constellation_running_time_list_dict = json.load(f)
```

```python
# plot a boxplot for each device
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
line_width = 2
font_size = 28
ax[0].boxplot(device_average_inference_time_list_dict.values(), labels=device_average_inference_time_list_dict.keys(), widths=0.5, boxprops=dict(linewidth=line_width), whiskerprops=dict(linewidth=line_width), capprops=dict(linewidth=line_width), medianprops=dict(linewidth=line_width))
# ax[0].xticks(fontsize=font_size)
ax[0].set_xticklabels(device_average_inference_time_list_dict.keys(), fontsize=font_size)
ax[0].set_xlabel('Device', fontsize=font_size)
# set y label and move it lower a bit
ax[0].set_ylabel('Inf. (ms)', fontsize=font_size)
ax0_yticks = [int(y) for y in ax[0].get_yticks()]
ax[0].set_yticklabels(ax0_yticks, fontsize=font_size)
ax[0].grid(axis='y')

# plot a boxplot for each constellation
ax[1].boxplot(constellation_running_time_list_dict.values(), labels=constellation_running_time_list_dict.keys(), widths=0.5, boxprops=dict(linewidth=2), whiskerprops=dict(linewidth=2), capprops=dict(linewidth=2), medianprops=dict(linewidth=2))
ax[1].set_xticklabels(constellation_running_time_list_dict.keys(), fontsize=font_size)
ax[1].set_xlabel('Constellation', fontsize=font_size)
ax[1].set_ylabel('Run. (s)', fontsize=font_size)
ax1_yticks = [int(y) for y in ax[1].get_yticks()]
ax[1].set_yticklabels(ax1_yticks, fontsize=font_size)
ax[1].grid(axis='y')

plt.tight_layout()
plt.savefig('fig6b_algorithm_and_model_time.pdf')
plt.savefig('fig6b_algorithm_and_model_time.png')
```
